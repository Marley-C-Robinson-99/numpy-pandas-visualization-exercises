```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as m
import seaborn as sns
```

```python
irs = sns.load_dataset('iris')
irs.species
```

```python
plt.boxplot(irs.petal_length)
```

```python
sns.relplot(x = "petal_length", y = "petal_width", data = irs)
```

```python
def avg_sepal_len(sepal_length):
    return irs.apply(avg(sepal_length))

def avg_sepal_width(sepal_width):
    return irs.apply(avg(sepal_width))
```

```python
sns.catplot(x = 'sepal_width', y = 'sepal_length', hue ='species', data = irs)
sns.relplot(x = 'sepal_length', y = 'sepal_width', hue ='species', data = irs)
```
