```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as m
```

```python
x = pd.Series(range(1,4))
```

```python
y0 = 0*x + 1
y1 = np.log(x)
y2 = x
y3 = x * (np.log(x))
y4 = x ** 2
y5 = 2**x
y6 = x.apply(m.factorial)
y7 = x**x
```

```python
plt.title('Big O Notation')
plt.xlabel("Elelments")
plt.ylabel('Operations')

plt.plot(x,y0, label = "$O(x)$")
plt.plot(x,y1, label = "$O(log_{x})$")
plt.plot(x,y2, label = "$O(x)$")
plt.plot(x,y3, label = "$O(x log_{x})$")
plt.plot(x,y4, label = "$O(x^2)$")
plt.plot(x,y5, label = "$O(2^x)$")
plt.plot(x,y6, label = '$O(x!)$')
plt.plot(x,y7, label = '$O(x^x)$')
plt.legend()
```
