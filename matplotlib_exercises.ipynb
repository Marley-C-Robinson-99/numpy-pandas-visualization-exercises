```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as m
```

```python
x = pd.Series(range(0,10))
def sqr_min_x_pls_2(x):
    return (x**2) -x + 2
y = x.apply(sqr_min_x_pls_2)
y.head()
```

```python
plt.plot(x, y, label = 'y=x^2 - x + 2')
```

```python
def squart(x):
    return m.sqrt(x)

y1 = x.apply(squart)
plt.plot(x, y1)
plt.title('√x')

plt.show()
```

```python
y2= x.apply(lambda x: x**3)
plt.plot(x,y2)
plt.title('x^3')
```

```python
y3= x.apply(lambda x: x*2)
plt.plot(x,y3)
plt.title('X*2')
```

```python
y4= x.apply(lambda x: 1/(x+1))
plt.plot(x,y3)
plt.title('1/X + 1')
```

```python
y1 = x.apply(squart)
plt.subplot(2,2,1)
plt.plot(x, y1)
plt.title('√x')

y2= x.apply(lambda x: x**3)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('x^3')

y3= x.apply(lambda x: 2**x)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('2^x')

y4= x.apply(lambda x: 1/(x+1))
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('1/X + 1')

plt.tight_layout()
```

```python
x = pd.Series(range(0,4))
y1 = x.apply(squart)
plt.plot(x, y1, 'g', label = '√x')
plt.title('√x')

y2= x.apply(lambda x: x**3)
plt.plot(x,y2, 'b',label = 'x**3')
plt.title('x^3')

y3= x.apply(lambda x: 2**x)
plt.plot(x,y3, 'k',label = '2**x')
plt.title('2^x')

y4= x.apply(lambda x: 1/(x+1))
plt.plot(x,y4, 'y',label = '1/(x+1)')
plt.title('Equations of X')

plt.legend(loc='upper left')
```
